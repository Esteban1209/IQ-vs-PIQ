#1. Importar librerias

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_palette('colorblind')
sns.set(style="whitegrid")
```

#2. Carga y exploracion de datos

```python
df = pd.read_csv('sample_data/brain_size.csv')
```

```python
print(df.head(5))
```

```python
print(df.describe())
```

#3. Limpieza de datos

Validacion de datos

```python
print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python
print(df.dtypes)
```

Limpieza

```python
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
print("Valores nulos: ", df.isnull().sum())
```

```python
print("Valores duplicados: ", df.duplicated().sum())
```

```python
print(df.dtypes)
```

Cambio de float a int

```python
df = df[["IQ", "PIQ"]].astype('Int64')
```

Validacion de los datos transformados

```python
print("\nValores nulos: ", df.isnull().sum())
print("\nValores duplicados: ", df.duplicated().sum())
print(df.dtypes)
df['IQ'] = df['IQ'] / 10
```

#4. Visualizaci칩n de Datos

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PIQ', y='IQ', data=df)
plt.title('Relaci칩n entre el tama침o del cerebro y el IQ')
plt.xlabel('Tama침o del cerebro)')
plt.ylabel('IQ')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['IQ'], kde=True)
plt.title('Distribuci칩n del CI')

plt.subplot(1, 2, 2)
sns.histplot(df['PIQ'], kde=True)
plt.title('Distribuci칩n del PIQ')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(y=df['PIQ'])
plt.title('Distribuci칩n de PIQ')
plt.ylabel('Puntuaci칩n PIQ')

plt.subplot(1, 2, 2)
sns.boxplot(y=df['IQ'])
plt.title('Distribuci칩n de IQ')
plt.ylabel('Puntuaci칩n IQ')

plt.tight_layout()
plt.show()
```

#5. An치lisis e Interpretaci칩n

```python
corr_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Matriz de Correlaci칩n')
plt.show()
```

Existe una fuerte correlacion entre las variables

Se observa una relaci칩n positiva entre ambas variables, a medida que aumenta el PIQ, tiende a aumentar tambi칠n el IQ. Esto sugiere que un mayor desempe침o en habilidades espec칤ficas podr칤a estar asociado con un IQ m치s alto.
